```python
grads = [92,88,43,11,33,22,87,50]
counter = 0 # index for access to item
pass_grads = []
while grads[counter]>50:
  pass_grads.append(grads[counter])
  counter +=1

print(pass_grads)
```

```python
pass_grads = []

for item in grads:
  if item>50:
    pass_grads.append(item)

print(pass_grads)
```

```python
def split_eshraq(text_e,split_e =" "):
  current_word =""
  words = []
  for char in text_e:
      if char == split_e:
          if current_word:
              words.append(current_word)
              current_word =""
      else:
          current_word += char

  if current_word:
      words.append(current_word)
  return words
```
