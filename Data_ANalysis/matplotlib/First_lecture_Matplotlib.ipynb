#Line plot

```python
#Line plot for time serie/ matmatical function / continues interval
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100) # linespace, the gap between it, is static
y = np.sin(x)
# represent between x and y(based on x )
# sin fun -> the input x, the output y
# f(x) = sin(x) = y
plt.plot(x, y)
plt.xlabel("this for x axis")
plt.ylabel("this for y axis")
plt.title("the title of plot, so line plot")
plt.grid(False)
plt.show()
```

#scatter plot

```python
# using to numical data, and showing the correlations/ realation as line(incress)
# and as Std for specific verabile based on another verabile
# as dot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.xlabel("this for x axis")
plt.ylabel("this for y axis")
plt.title("the title of plot, so Scatter")
plt.grid(True)
plt.show()
```

# Histogram plot

```python
# statical plot -> to know the dist of one value
# unvarite / one verabile
# noraml dist/ Gauss distribution - the relation is around median -> normal
# It divides the data into intervals (bins),
# and the height of the bars represents the number of data points in each bin.
```

```python
# data = np.array([100, 30, 20, 50, 50, 60, 55, 66, 10, 90, 80] * 5)
data = np.random.randn(1000)
# data = np.random.rand(1000) # wrong bec , the histogram for normal dist ,around median

plt.hist(data, bins = 40, edgecolor="pink")
plt.xlabel("marks")
plt.ylabel("Frequency")
plt.title("the title of plot, so hist")
plt.grid(False)
plt.show()
```

# Pie Charts
Pie charts are circular statistical graphics divided into slices to illustrate numerical proportion.

```python
# Show as percentage of  diff category
```

```python
# Sample data
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]

# Create pie chart
plt.pie(labels=labels, x=sizes, autopct='%1.1f%%', startangle=270)
# Positional
plt.title('Pie Chart Example')
plt.show()
```


# Bar Charts

Bar charts are used to display quantities corresponding to different categories.
They consist of rectangular bars, where the length of the bar represents the quantity of the category.


```python
# Show as numbers of  diff category
```

```python
categories = ['A', 'B', 'C', 'D']
values = [4, 7, 1, 8]

# Create bar chart
plt.bar(height=values, x=categories) # non Positional
plt.title('Bar Chart Example')
plt.xlabel('Category')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```
