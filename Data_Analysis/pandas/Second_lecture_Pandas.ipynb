#Concatenating

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'W': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 5, 1, 3]
                   )
#df1.reset_index(drop=True)
```

```python
df2 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7'],
                        'F': [0, 0, 0, 0]},
                         index=[8, 9, 4, 2]
                   )
```

```python
df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        #index=[8, 9, 10, 11]
                   )
```

```python
pd.concat([df1, df2, df3], ignore_index=True, sort=True)
# index ->
# if even defult or get vaules for indeex
# you give it numbers from 0
# sort -> sort the columns
# the index important (meaning full) when you have relations with values
```

```python
pd.concat([df1, df2, df3], ignore_index=True, sort=False, join="outer")
```

```python
pd.concat([df1, df2, df3], ignore_index=True, sort=True, join="outer")
```

```python
pd.concat([df1, df2], axis=0, verify_integrity=True)
# for ensure no duplicate in column if axis = 1,
# no duplicate in indices if axis = 0,
```

```python
import pandas as pd

data = {'A': [1, 3.4, 3], 'B': [4.5, 5, 6]}
data1 = {'c': [4.4, 5.4, 6.4], 'Q': [14.5, 5, 6]}

df = pd.DataFrame(data)
df1 = pd.DataFrame(data1)

pd.concat([df, df1],axis = 1, verify_integrity=True)
```

```python

data = {'A': [1, 3.4, 3], 'B': [4.5, 5, 6]}
data1 = {'C': [4.4, 5.4, 6.4], 'D': [1.4, 3.3, 33.3]}

df = pd.DataFrame(data)
df1 = pd.DataFrame(data1)

# Remove verify_integrity or set it to False to allow concatenation with duplicate indices
pd.concat([df, df1], ignore_index=True)
```

```python
import pandas as pd
students = pd.DataFrame({'student_name': ['Shadi', 'Samer', 'Walaa', 'Nuha'],
                     'age': ['34', '23', '24', '31'],
                     'course': ['AI', 'AI', 'Business', 'CS']})

courses = pd.DataFrame({'course': ['AI', 'Business', 'CS', 'BI'],
                          'instructor': ['Mr. Waleed', 'Ms. Samar', 'Mr. Ahmad', 'Ms. Batool']},)
                          # 'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
students
```

```python
courses
```

```python
pd.merge(left=students, right=courses, how="inner", left_on='course', right_on='course_name')
```

```python
pd.merge(left=students, right=courses, how="inner", on="course")
```

```python
pd.merge(left=students, right=courses, how="outer", on="course")
```

```python
pd.merge(left=students, right=courses, how="right", on="course")
```

```python
pd.merge(left=students, right=courses, how="left", on="course")
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']},
                         index = [1, 2, 3, 4]
                         )

right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']},
                                   index = [1, 2, 5, 4]
                     )

```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how="inner", left_index=True, right_index=True)
# based on common indices
```

```python
pd.merge(left, right, how="inner", on ="key1") # based on key1
```

```python
pd.merge(left, right, how="inner")
```

```python
pd.merge(left, right, left_on="key1", right_on="key1" )
```

```python
pd.merge(left, right, on =["key1", "key2"])
```

```python
pd.merge(left, right,how="outer", on=["key1", "key2"])
```

```python
pd.merge(left, right,how="left", on=["key1", "key2"])
```

```python
left
```

```python
# cross -> between full 2 dataFrame -> product
pd.merge(left, right, how="cross")
# cross: creates the cartesian product from both frames,
# preserves the order of the left keys.
```

```python
pd.merge(left=left, right=right, left_on="key1", right_on="key2")
```

#joining

```python
# is method from instance of class
# for simple operations
# Joining is a convenient method for
# combining the columns of two potentially differently-indexed DataFrames
# into a single result DataFrame
# join as adding
# join based on one column
```

```python
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3'],
                     'A': ['A0', 'A1', 'A2']},
                      index=['K0', 'K2', 'K3'])

print(left)
print(right)
```

```python
left
```

```python
right
```

```python
left.join(right)
```

```python
left.join(right, how="right")
```

```python
right.join(left)
```

```python
left.join(right, how='outer', lsuffix='ee', rsuffix='ss')
```
