```python
# Index Hierarchy or multi-Index
# which is an iterator of (tuples)
# where the first item in each passed iterator is paired together,
# and then the second item in each passed iterator are paired together
outside = ["it", "it", "it", "hr", "hr", "hr"]
inside =  ["ai", "cs", "cis", "hr1", "hr2", "hr3"]
#The zip() function returns a zip object
hier_index = list(zip(outside, inside))
                # tuple as a pair
print(hier_index)

hier_index = pd.MultiIndex.from_tuples(hier_index)
```

#Groupby

```python
#  Groupby as obj
import pandas as pd
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# Now you can use the .
# groupby() method to group rows together based off of a column name.
dfe = df.groupby("Person") # as an object #DataFrameGroupBy object
# and groupby, reordering
```

```python
df.groupby("Person")
```

```python
#  aggregate methods
 # mean, min, max, std -> for numerical columns
dfe.mean(numeric_only=True)# in the numerical columns
```

```python
df1 = df.groupby("Company")
# eash company as one group, one row
df.groupby("Company").min(numeric_only=True)
```

```python
df1.min(numeric_only=True)
```

```python
df.groupby("Company").max(numeric_only=True)
```

```python
df.groupby("Company").std(numeric_only=True)
```

```python

df.groupby("Company").count()
```

```python
df1.describe()
```

```python
df1.describe().transpose()["FB"]
```

```python
df.groupby("Company").describe()
```
