#pivotTable

```python
#      
```

```python
data = {'A':['foo','foo','foo','bar','bar','bar'],
     'B':['one','one','one','two','one','one'],
       'C':['x','x','x','y','x','y'],
       'D':[11,3,2,5,4,1]}

df = pd.DataFrame(data)
```

```python
df
```

```python
pd.pivot_table(df, index=["A", "B"], columns="C", values="D")
#foo	one	7.0	NaN  -> 7.0 is a avg bec the aggfunc='mean'
```

```python
pd.pivot_table(df, index=["A", "B"], columns="C", values="D", aggfunc='sum')
```
