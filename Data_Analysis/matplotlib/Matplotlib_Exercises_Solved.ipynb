```python
import numpy as np
x = np.arange(0,100) # from 0 to 99
y = x*2 # linear relationship bet x and y -> y = 2x+0
z = x**2 # y = ax^2+bx^1+c, the large power  is 2
# v = x^3 -> the large power is 3 -> ax^3+bx^2+dx^1+c
```

## Exercise 1

** Follow along with these steps: **
* ** Create a figure object called fig using plt.figure() **
* ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
                 #[left, bottom, width, height]
ax.plot(x, y, color="#0000FF")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("title")
```

## Exercise 3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
axis_1 = fig.add_axes([0,0,1,1])
axis_2 = fig.add_axes([0.2,0.5,.4,.4]);
```

** Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
color_1 = "#0000FF"
axis_1.plot(x, z, color=color_1)
axis_1.set_xlabel("X")
axis_1.set_ylabel("Z")

axis_2.plot(x, y, color=color_1)
axis_2.set_xlabel("X")
axis_2.set_ylabel("Y")
axis_2.set_title("zoom")
axis_2.set_xlim([20, 22]) # limit
axis_2.set_ylim([30, 50])
fig
```
