```python
import numpy as np
x = np.linspace(-10, 10, 20)
y = x ** 2 # this function f(x) = x^2 =y
plt.plot(x, y, 'red') #  'r' is the color red
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
#plt.plot(y, y, 'g')
plt.show()
```

```python
# Creating Multiplots on Same Canvas
# figure here as a matrix
# dim1 of row(from 1), dim2 of col(from 1), index (existing) from (1)
# dim1 * dim2 = # of plot
plt.subplot(2,1,1)
plt.plot(x, y, 'r--') # is line style
plt.subplot(2,1,2)
plt.plot(y, x, 'g*-')
plt.show()
```

#Matplotlib Object Oriented Method

```python
# line plot
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
# plot is a finger and axis (objs)
```

```python
# to show specific thing in a main plot on the inst plot
fig_1 = plt.figure() # this obj , as white/blank paper
axis_1 = fig_1.add_axes([0, 0, 0.9, 2]) # left, bottom, width, height (range 0 to 1)
# axis_1 is obj , axis_1 based on the blank paper in the axis
axis_1.plot(x, y,"y*-")
# axis contain all data/axis/title -> elemant
axis_1.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axis_1.set_ylabel('Set y Label')
axis_1.set_title('Set Title')
plt.show()
```

```python
fig_2 = plt.figure()

axis_1 = fig_2.add_axes([0, 0, 0.8, 0.8]) # main axis, below inset axis
# axis_1 based on the blank paper in the axis
axis_1.plot(x, y,"y*-")
axis_1.set_xlabel('Set X Label_1')
axis_1.set_ylabel('Set y Label_1')
axis_1.set_title('Set Title_1')
plt.grid()

                  #0.20% from left, #0.40% from bottom
axis_2 = fig_2.add_axes([0.2, 0.4, 0.2, 0.2])  # left, bottom  as percentage
                                                # based on axis_1
# axis_2 based on the axis_1 in the axis
axis_2.plot(y, x,"p")
axis_2.set_xlabel('Set X Label_2')
axis_2.set_ylabel('Set y Label_2')
axis_2.set_title('Set Title_2')
plt.grid();
```

# subplots

```python
%matplotlib inline
# fig, axes = plt.subplots() this 1 * 1 so one axis not list of array
# is normal axis
fig , axis = plt.subplots(1, 2, sharex=True, sharey=True)
# sharey, x to compare between together things
# (as level the AI and CS student depending on Calculace(static as y_label))
# axis  = (axis_1, axis_2) is unpaking
# axis_1.plot(y, x, "g*-")
axis[0].plot(y, x, "g*-")
axis[0].plot(x, x)
axis[0].set_xlabel("is x")
axis[0].set_ylabel("is y")
axis[0].set_title("the first axis")

axis[1].plot(x, y, "b*")
axis[1].set_xlabel("is y")
axis[1].set_ylabel("is x")
axis[1].set_title("the second axis")
# plt.tight_layout()
# so if overlapping exit in subplots or figures.
# We ca use fig.tight_layout() or plt.tight_layout() method,
# which automatically adjusts the positions of the axes on the figure
# to illustration plots
fig.tight_layout()

plt.show()
```
