#matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
                # sub module
# plt is module
# all the graph in the same plot
x_axis = np.arange(1, 10, 2)
y_axis = np.exp(x_axis) / 100
# the value of exp = (approximately 2.71828) e to power(1, 10, 2)
```

```python

plt.title("graph")
plt.xlabel("x") # the name of x_axis
plt.ylabel("y") #the name of y_axis
plt.plot(x_axis, y_axis ) #the type of plot is  solid line
#plot is fun
plt.grid(True) # for line in background
plt.show()
```

```python
# بيانات المخطط الأول
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]

# بيانات المخطط الثاني
x2 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]

# رسم المخطط الأول
plt.plot(x1, y1, label="Squared", color="blue")

# رسم المخطط الثاني
plt.plot(x2, y2, label="Linear", color="green")

# إضافة عنوان وتسمية المحاور
plt.title("Multiple Graphs on the Same Plot")
plt.ylabel("Y Axis")
plt.xlabel("X Axis")
plt.legend() # the lebel to show the lebel argument

plt.show()
```

```python
# other types of plots
# sctter, plot, bar
import matplotlib.pyplot as plt
import numpy as np

# إنشاء figure ومحاور (1 صف و 2 أعمدة)
# fig is obj
# axes is array of obj
fig, axes = plt.subplots(2, 1)

# البيانات للمخططات
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# رسم المخطط الأول على المحور الأول
axes[0].scatter(x, y1, label = "test1", c="red")
axes[0].set_title("Sine Wave")
axes[0].set_xlabel("X")
axes[0].set_ylabel("Sin(X)")
axes[0].legend()

# رسم المخطط الثاني على المحور الثاني
axes[1].plot(x, y2, label = "test2")
axes[1].set_title("Cosine Wave")
axes[1].set_xlabel("X")
axes[1].set_ylabel("Cos(X)")
axes[1].legend()

# ضبط التخطيط بحيث تكون الرسوم غير متداخلة
plt.tight_layout()

# عرض الرسم
plt.show()
```
