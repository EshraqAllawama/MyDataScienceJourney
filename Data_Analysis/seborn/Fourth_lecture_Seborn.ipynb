```python
pldata = sns.PairGrid(data_1)
d = pldata.map(sns.kdeplot)
plt.show(d)
```

```python
pldata = sns.PairGrid(data_1)
# kernel density estimation
# 1 varite -> 2d -> x: data, y: density (depth)
# 2 varite -> 3d as 2 d -> x: data, y: data, z virtual: density (depth)
pldata.map_diag(sns.kdeplot) # one varit
pldata.map_upper(sns.kdeplot) # two varit
pldata.map_lower(plt.scatter)
#pldata.map_diag(plt.hist)

plt.show()
```

```python
pldata = sns.PairGrid(data_1)

pldata.map_offdiag(sns.kdeplot)
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# as cut data into subsets -> as sex=male|smaoker=yes
g = sns.FacetGrid(tips, row='sex', col='smoker', hue='time') #layout, white paper
g.map(sns.kdeplot, 'total_bill')
# g.map(sns.kdeplot, 'size')  kde -> Adjusts to categories

plt.legend()
```

```python
g = sns.FacetGrid(tips, row='sex', col='smoker', hue='time') #layout, white paper

g.map(plt.scatter, "total_bill", "tip").add_legend()
```

```python
tips
```

```python
#jointplot
g = sns.JointGrid(x='total_bill', y='tip', data=tips)
g.plot(sns.kdeplot, sns.kdeplot)
```

```python
g = sns.JointGrid(x='total_bill', y='tip', data=tips, hue='time')
g.plot(sns.histplot, sns.boxplot)
```

```python
g = sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
g = sns.jointplot(x='total_bill', y='tip', data=tips)

sns.histplot(x=tips['total_bill'], fill=False, linewidth=2, ax=g.ax_marg_x).plot
#sns.histplot(x=tips['total_bill'], fill=False, linewidth=2, ax=g.ax_marg_y)
```
