```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
sns.countplot(x='sex',data=tips, palette='Set2')
```

```python
sns.set_style('darkgrid')
```

```python
sns.set_style('ticks')
```

```python
# bet catogiral and numerical
sns.lineplot(x='sex', y='total_bill', data=tips)
#sns.lineplot(x='sex', y='total_bill', data=tips, palette='deep', hue='sex')
```

```python
sns.despine(top=True, right=True, left=True, bottom=True)
sns.countplot(x='sex',data=tips)
plt.show()
```

You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect

```python
plt.figure(figsize=(20,4)) # W and H
sns.countplot(x='sex',data=tips)
```

interface to fit regression models across conditional subsets of a dataset.

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex',markers=['o','v'], aspect=2) # aspect ratio
```
