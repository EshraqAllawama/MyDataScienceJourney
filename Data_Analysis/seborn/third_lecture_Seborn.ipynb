#clustering
The clustermap uses hierarchal clustering to produce a clustered version of the heatmap.
##groping use heatmap

```python
sns.clustermap(flights_test, linecolor="white", linewidths=2, annot=True, annot_kws={'size':8})
# بحط عندي ال 2 v
# المنتشابهه بقروب واحد
# as sibling  بكون في علاقة
```

Notice now how the years and months are no longer in order, instead they are grouped by similarity in value (passenger count). That means we can begin to infer things from this plot, such as August and July being similar (makes sense, since they are both summer travel months)

standard_scale=0: يعمل standardization على مستوى الصفوف (rows).

standard_scale=1: يعمل standardization على مستوى الأعمدة (columns)

لما بيتقالك "standardize that dimension" (يعني توحيد البعد)، المقصود هنا هو عملية تهيئة (normalization) للبيانات على مستوى كل صف (row) أو عمود (column) في المصفوفة بتاعتك. الطريقة دي بتتضمن خطوتين:

طرح القيمة الدنيا (subtract the minimum): بتاخد أقل قيمة في الصف أو العمود وتطرحها من كل القيم في نفس الصف أو العمود.
القسمة على القيمة العظمى (divide by the maximum): بعد ما تطرح القيمة الدنيا، بتقسم كل قيمة على أكبر قيمة موجودة في نفس الصف أو العمود (بعد الطرح)

```python
# More options to get the information a little clearer like normalization ->in specific range (نسبة وتناسب)
# coolwarm is more clear
sns.clustermap(flights_test, cmap="coolwarm", linecolor="white", linewidths=2, annot=True, standard_scale=0)
# the biggest number is 1 likes 600
# نسبة وتناسب بين اكبر رقم مع واحد, وناسب باقي الارقام لاقل رقم مع 0
```
