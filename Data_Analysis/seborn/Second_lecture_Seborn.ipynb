#barplot and countplot

These very similar plots allow you to get aggregate data off a categorical feature in your data. **barplot** is a general plot that allows you to aggregate the categorical data based off some function, by default the mean:

```python
# for bi var (math fun for numrical var)
# calaulte the mean for each cat
sns.barplot(x='sex', y='tip', data=data_1 )
# حسب متوسط البقشييش لكل cat لحال
```

#boxplot

boxplots and violinplots are used to shown the distribution of categorical data. A box plot (or box-and-whisker plot) shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
sns.boxplot( x="total_bill", y="day", data=data_1, palette='rainbow', orient= 'h')
# the boxs by defult horzantle

# expalin for all
# المحل معطل الاثنين والثلااء والاربعاء
# 1- sat : 25 من الفواتير بيوم السبت اقل من 15 dollar
# 1- sat : نص الفواتير في يوم ال سبت اقل من 19 دولار بالنسبة للفواتير بهاذ اليوم
# 2- كل ما كان البوكس مضبضب اكثر معناتو تركيز الداتا اكثر معناتو في std معناتو نرمال توزيع
# وال iqr كل ما زاد معناتو في std اكثر معناتو في سكواد وبطل نورمال
# 3-والوسكي بلوت الطويلة معناها في داتا مبعثرة كثير,
# error margen الاختلاف بين شغلي البردكت والواقع الفعلي بس, اقل او اعلى بنسبة من الواقع
# 4- في عندي اكمن شخص قلال فاتورتهم تجاوزت الاربعين لحد 33 يوم الثلاثاء
# التباين : انتشار الداتا او يسمى ال varince , كل ما زاد زاد ال std وزاد نتشار الداتا وبطلت مركزة
```

#violin plot
A violin plot plays a similar role as a box and whisker plot. It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that those distributions can be compared (اذا بدي اعمل بردكت ليوم معين لازم يكون في اليوم الثاني سبشل او مخالف له, يكون التنبؤ عندي بحالة ممتازة, لتميز يعني). Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution

```python

# kde and boxplot
sns.violinplot(x="day", y="total_bill", data=data_1, palette='rainbow')
```

```python
#dist 3
sns.violinplot(x="day", y="total_bill",hue='sex', data=data_1, order = ["Sun","Thur", "Fri", "Sat"], palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill",hue='sex',split = True,hue_order=['Female','Male'], data=data_1, palette='rainbow')
```

#stripplot
The stripplot will draw a scatterplot where one variable is categorical. A strip plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.

```python
# as scatter
# bet cat and num as scatter
sns.stripplot(x='day', y='total_bill', data=data_1)
```

```python
sns.stripplot(x='day', y='total_bill', data=data_1, jitter=False) # True for explain data and distrbute it
```

```python
sns.boxplot(x='day', y='total_bill', data=data_1, palette='coolwarm', hue='sex')
sns.stripplot(x='day', y='total_bill', data=data_1, palette='coolwarm', hue='sex')
```

#swarmplot
The swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This gives a better representation of the distribution of values, although it does not scale as well to large numbers of observations (both in terms of the ability to show all the points and in terms of the computation needed to arrange them).

```python
# as sccater
# avoid overlaping( تجمهر الداتا ع بعض - تداخل) by defult
# above with prammaters
```

```python
sns.swarmplot(x='day', y='total_bill', data=data_1)
```

```python
sns.swarmplot(x='day', y='total_bill', data=data_1, palette='Set1', hue='size', legend=False, native_scale=False)
```

```python
sns.swarmplot(x='size', y='total_bill', data=data_1, palette='Set1', hue='size', native_scale=False)
```

```python
sns.swarmplot(x='size', y='total_bill', data=data_1, palette='Set1', hue='size', native_scale=True)
```
