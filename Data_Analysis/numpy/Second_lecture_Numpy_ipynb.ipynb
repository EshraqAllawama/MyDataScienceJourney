#Fancy Indexing
Fancy indexing allows you to select entire rows or columns out of order,

```python
# Select Multiple Elements Using NumPy Fancy Indexing
arr2d = np.zeros((10,10))
```

```python
lenght_of_array = arr2d.shape[0]
```

```python
for num in range(lenght_of_array):
    for num1 in range(lenght_of_array):
        arr2d[num,num1] = num1
```

```python
arr2d
```

```python
arr2d[::2, 3:7]
```

```python
arr2d[[0, 2, 4, 6, 8], 3:7]
```

```python
arr2d = np.arange(100).reshape(10,10)
```

```python
arr2d
```

```python
arr2d[[0, 2, 4, 6, 8], 3:7]
```

```python
#Allows in any order
arr2d[[4, 0, 6, 8, 2], 3:7]
```
